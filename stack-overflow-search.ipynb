```python
import html
import itertools
from urllib import parse

import requests
```

```python
url = 'https://api.stackexchange.com/2.2/questions'
query = {
    'page': '1',
    'pagesize': '100',
    'order': 'desc',
    'sort': 'votes',
    'tagged': 'bash',
    'site': 'stackoverflow'
}
```

```python
response = requests.get(url, params=query)
questions = response.json()['items']
```

```python
[(q["score"], q["title"]) for q in itertools.islice(questions, 10)]
```

```python
links = []

for n, q in enumerate(questions, start=1):
    other_tags = " ".join(
        f"``[{t}]``" for t in sorted(q["tags"]) if t != "bash"
    )
    title = html.unescape(q['title'])
    print(f"{n:d}. {q['score']:>+4d} `{title}`_\n    {other_tags}\n")
          
    links.append(f".. _{title}: {q['link']}")
#     if n == 10: break
```

```python
print("\n".join(links))
```
